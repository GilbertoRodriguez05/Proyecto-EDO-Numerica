Tema 9 A     Campo de Isoclinas

```python
import numpy as np
import matplotlib.pyplot as plt

# Parámetros
k = 1  # constante de crecimiento

# Rango de valores
t = np.linspace(0, 300, 100)   # eje del tiempo
P = np.linspace(0, 300, 20)   # eje P (variable dependiente)
T, Pm = np.meshgrid(t, P)    # malla

# Ecuación diferencial: dP/dt = k * sqrt(P)
dP = k * np.sqrt(Pm)
dt = np.ones_like(dP)  # componente horizontal

# Normalizar los vectores para que tengan tamaño uniforme visualmente
norm = np.sqrt(dt**2 + dP**2)
dt /= norm
dP /= norm

# Función solución
def function_solved(t, P0):
    return (k*t/2 + np.sqrt(P0))**2
images1 = function_solved(t, 100)
images2 = function_solved(t, 16)
images3 = function_solved(t, 4)

# Graficar campo de vectores (isoclinas)
plt.figure(figsize=(12, 8))
plt.quiver(T, Pm, dt, dP, angles='xy', color='teal', width=0.0015, scale=30)

# Distinos casos de valores iniciales
plt.plot(t, images1,"r-", linewidth=2, label= "P0=100")
plt.plot(t, images2,"b-", linewidth=2, label= "P0=16")
plt.plot(t, images3,"k-", linewidth=2, label= "P0=4")

plt.title(r'Campo de Isoclinas de $\frac{dP}{dt} = k\sqrt{P}$', fontsize=14)
plt.xlabel('Tiempo (t)')
plt.ylabel('P')
plt.grid(alpha=0.3)
plt.xlim(0, 50)
plt.ylim(0, 300)
plt.legend()
plt.show()
```
